```python
import pandas as pd

org_pub = "pub.csv"
```

```python
df = pd.read_csv(org_pub, index_col=False)
```

```python
auth_list = []

for index, item in df.iterrows():
    count = 1
    auth_to_pub_map = {}

    auth_list_name = item["auth_name"]
    
    if "," in auth_list_name:
        split_list = auth_list_name.split(",")
        
        for auth in split_list:
            auth1 = auth.strip()
            index, val = find(auth_list, "auth", auth1)
            if index == -1:
                auth_to_pub_map["auth"] = auth1
                auth_to_pub_map["count"] = count
                auth_list.append(auth_to_pub_map)
            else:
                auth_list[index]["count"] = val + 1
            
    else:
        auth1 = auth_list_name.strip()
        index, val = find(auth_list, "auth", auth1)
        if index == -1:
            auth_to_pub_map["auth"] = auth1
            auth_to_pub_map["count"] = count
            auth_list.append(auth_to_pub_map)
        else:
            auth_list[index]["count"] = val + 1

    
auth_list
```

```python
def find(lst, key, value):
    for i, dic in enumerate(lst):
        if key in dic and dic[key] == value:
            return i, dic["count"]
    return -1, -1
```

```python
auth_sorted_list = sorted(auth_list, key=lambda d: d['count'], reverse=True) 
```
